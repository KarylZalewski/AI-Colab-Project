```python
import numpy as np
from sklearn.linear_model import Perceptron
import matplotlib.pyplot as plt

############# ZBIÓR UCZĄCY ####################
X = np.array([[0],[1]]) #wektor cech (wejścia)#
Y = np.array([1,-1]) #wektor klas (wyjścia)   #
###############################################

# TO DO:
# Stworzyć perceptron
perceptron = Perceptron()
# Nauczyć perceptron
# Wypisać wartości wag i biasu
# tutaj ta pętla jest w zasadzie nie potrzebna
perceptron.fit(X,Y,coef_init=None, intercept_init=None, sample_weight=None)
print(perceptron.coef_, perceptron.intercept_)
# Narysować liniową funkcję klasyfikującą dla nauczonego perceptronu
funkcja = (perceptron.coef_*X + perceptron.intercept_)
# proszę jeszcze na tym wykresie zaznaczyć zbiór uczący i punkt, który rozdziela
# dwie klasy ze zbioru uczącego
x_point = 0.5
y_point = 0
plt.scatter(x_point,y_point)
plt.xlabel("Zbior uczacy")
plt.plot(funkcja)
plt.show()
```

```python
#projekt na wakacje
import numpy
import sys
from nltk.tokenize import RegexpTokenizer
from nltk.corpus import stopwords
from keras.models import Sequential
from keras.layers import Dense, Dropout, LSTM
from keras.utils import np_utils
from keras.callbacks import ModelCheckpoint
file = open("frankenstein-2.txt").read()
def tokenize_words(input):
    # lowercase everything to standardize it
    input = input.lower()

    # instantiate the tokenizer
    tokenizer = RegexpTokenizer(r'\w+')
    tokens = tokenizer.tokenize(input)

    # if the created token isn't in the stop words, make it part of "filtered"
    filtered = filter(lambda token: token not in stopwords.words('english'), tokens)
    return " ".join(filtered)
processed_inputs = tokenize_words(file)
chars = sorted(list(set(processed_inputs)))
char_to_num = dict((c, i) for i, c in enumerate(chars))
input_len = len(processed_inputs)
vocab_len = len(chars)
print ("Total number of characters:", input_len)
print ("Total vocab:", vocab_len)
seq_length = 100
x_data = []
y_data = []
# loop through inputs, start at the beginning and go until we hit
# the final character we can create a sequence out of
for i in range(0, input_len - seq_length, 1):
    # Define input and output sequences
    # Input is the current character plus desired sequence length
    in_seq = processed_inputs[i:i + seq_length]

    # Out sequence is the initial character plus total sequence length
    out_seq = processed_inputs[i + seq_length]

    # We now convert list of characters to integers based on
    # previously and add the values to our lists
    x_data.append([char_to_num[char] for char in in_seq])
    y_data.append(char_to_num[out_seq])
n_patterns = len(x_data)
print ("Total Patterns:", n_patterns)
X = numpy.reshape(x_data, (n_patterns, seq_length, 1))
X = X/float(vocab_len)
y = np_utils.to_categorical(y_data)
model = Sequential()
model.add(LSTM(256, input_shape=(X.shape[1], X.shape[2]), return_sequences=True))
model.add(Dropout(0.2))
model.add(LSTM(256, return_sequences=True))
model.add(Dropout(0.2))
model.add(LSTM(128))
model.add(Dropout(0.2))
model.add(Dense(y.shape[1], activation='softmax'))
model.compile(loss='categorical_crossentropy', optimizer='adam')
filepath = "model_weights_saved.hdf5"
checkpoint = ModelCheckpoint(filepath, monitor='loss', verbose=1, save_best_only=True, mode='min')
desired_callbacks = [checkpoint]
num_to_char = dict((i, c) for i, c in enumerate(chars))
start = numpy.random.randint(0, len(x_data) - 1)
pattern = x_data[start]
print("Random Seed:")
print("\"", ''.join([num_to_char[value] for value in pattern]), "\"")
for i in range(1000):
    x = numpy.reshape(pattern, (1, len(pattern), 1))
    x = x / float(vocab_len)
    prediction = model.predict(x, verbose=0)
    index = numpy.argmax(prediction)
    result = num_to_char[index]

    sys.stdout.write(result)

    pattern.append(index)
    pattern = pattern[1:len(pattern)]

```

```python
#Sieć GAN
import tensorflow as tf
import glob
import imageio
import matplotlib.pyplot as plt
import numpy as np
import os
import PIL
from tensorflow.keras import layers
import time
from IPython import display
import keras
import pathlib
dataset_url = "https://storage.googleapis.com/download.tensorflow.org/example_images/flower_photos.tgz"
archive = tf.keras.utils.get_file(origin=dataset_url, extract=True)
data_dir = pathlib.Path(archive).with_suffix('')
#dane = "/content/drive/MyDrive/Mushrooms/Amanita1" #"Agrocybe_aegerita" Amanita
tf.config.list_physical_devices('GPU')
#tf.debugging.set_log_device_placement(True)
s_px = 224
w_px = 224
rozmiar_probki = 150 #36
BUFFER_SIZE = 6000
BATCH_SIZE = 256
m = 7
n = 196 #28
s = 98 #14
r = 3 #1
#256
#(train_images, train_labels), (_, _) = tf.keras.datasets.mnist.load_data()
#(train_images, train_labels), (_,_)= tf.keras.datasets.mnist.load_data()
#dane = dane.reshape(dane.shape[0], 196, 196, 3).astype('float32')
#train_images = (train_images - 127.5) / 127.5
#train_dataset = tf.data.Dataset.from_tensor_slices(train_images).shuffle(BUFFER_SIZE).batch(rozmiar_probki)
#z_ucz = tf.keras.preprocessing.image_dataset_from_directory(dane,seed=66, image_size=(s_px,w_px),batch_size = rozmiar_probki,labels=None, color_mode='grayscale')
train_ds = tf.keras.utils.image_dataset_from_directory(
  data_dir,
  seed=123,
  labels=None,
  image_size=(s_px, w_px),
  batch_size=rozmiar_probki,
  validation_split=0.99,
  subset="training")
#print(z_ucz)
def normalize(image):
  return (image - 127.5)/127.5
train_ds = train_ds.map(normalize)
#normalization_layer = tf.keras.layers.Rescaling(1./127.5, offset=-1)
#normalized_ds = train_ds.map(lambda x: (normalization_layer(x)))
#image_batch = next(iter(normalized_ds))
#first_image = image_batch[0]
#print(np.min(first_image), np.max(first_image))
data_slice = 60000
#train_labels = keras.utils.to_categorical(train_labels)
#assert train_images.shape == (10000, 28, 28)
#assert train_labels.shape == (10000,)
#train_images = train_images[:data_slice,:]
#train_labels = train_labels[:data_slice,:]
#print(len(train_images),len(train_labels))
#train_images = train_images.reshape(train_images.shape[0], 28, 28, 1).astype('float32')
#train_images = (train_images - 127.5) / 127.5
#train_dataset = tf.data.Dataset.from_tensor_slices(train_images).shuffle(BUFFER_SIZE).batch(BATCH_SIZE)
#for image in train_ds.as_numpy_iterator():
  #print(np.amax(image))
  #print(np.amin(image))
#for image in z_ucz.as_numpy_iterator():
  #print(np.amax(image))
  #print(np.amin(image))
#z_ucz = (z_ucz - 255)/255
#z_test = tf.keras.preprocessing.image_dataset_from_directory(dane, validation_split=0.1, subset="validation", seed=66, image_size=(s_px,w_px), batch_size=rozmiar_probki)
def make_generator_model():
    model = tf.keras.Sequential()
    model.add(layers.Dense(m*m*256, use_bias=False, input_shape=(100,), activation='tanh'))
    model.add(layers.BatchNormalization())

    model.add(layers.Reshape((m, m, 256)))
    assert model.output_shape == (None, m, m, 256)
    model.add(layers.Conv2DTranspose(128, (16, 16), strides=(1, 1), padding='same', use_bias=False,))#activation='tanh'))
    #model.add(layers.UpSampling2D(input_shape=(n, n, r)))
    assert model.output_shape == (None, m, m, 128)
    #assert model.output_shape == compute_output_shape(input_shape)
    model.add(layers.BatchNormalization())
    model.add(layers.LeakyReLU())

    model.add(layers.Reshape((m, m, 128)))
    assert model.output_shape == (None, m, m, 128)
    model.add(layers.Conv2DTranspose(64, (32, 32), strides=(2, 2), padding='same', use_bias=False, )) #activation='tanh'))
    assert model.output_shape == (None, m*2, m*2, 64)
    model.add(layers.BatchNormalization())
    model.add(layers.Dropout(0.3))
    model.add(layers.LeakyReLU())

    model.add(layers.Reshape((m*2, m*2, 64)))
    assert model.output_shape == (None, m*2, m*2, 64)
    model.add(layers.Conv2DTranspose(64, (16, 16), strides=(1, 1), padding='same', use_bias=False))
    assert model.output_shape == (None, m*2, m*2, 64)
    model.add(layers.BatchNormalization())
    model.add(layers.LeakyReLU())

    model.add(layers.Reshape((m*2, m*2, 64)))
    assert model.output_shape == (None, m*2, m*2, 64)
    model.add(layers.Conv2DTranspose(32, (64, 64), strides=(2, 2), padding='same', use_bias=False,)) #activation='tanh')) #<= do przywrócenia
    assert model.output_shape == (None, m*4, m*4, 32)
    model.add(layers.BatchNormalization())
    #model.add(layers.LeakyReLU())

    #model.add(layers.Reshape((m*4, m*4, 32)))
    #assert model.output_shape == (None, m*4, m*4, 32)
    model.add(layers.Conv2DTranspose(16, (128, 128), strides=(2, 2), padding='same', use_bias=False)) #activation='tanh')) #<= do przywrócenia
    assert model.output_shape == (None, m*8, m*8, 16)
    model.add(layers.BatchNormalization())
    model.add(layers.LeakyReLU())

    #model.add(layers.Reshape((m*8, m*8, 16)))
    #assert model.output_shape == (None, m*8, m*8, 16)
    model.add(layers.Conv2DTranspose(8, (128, 128), strides=(2, 2), padding='same', use_bias=False,)) #activation='tanh')) #<= do przywrócenia
    assert model.output_shape == (None, m*16, m*16, 8)
    model.add(layers.BatchNormalization())
    model.add(layers.LeakyReLU())

    model.add(layers.Conv2DTranspose(3, (128, 128), strides=(2, 2), padding='same', use_bias=False, activation='tanh'))
    assert model.output_shape == (None, 224, 224, 3) # [224, 224] 1
    return model
generator = make_generator_model()
noise = tf.random.normal([1, 100])
generated_image = generator(noise, training=False)
plt.imshow(generated_image[0, :, :, 0], cmap='gray')
def make_discriminator_model():
    model = tf.keras.Sequential()
    model.add(layers.Conv2D(64, (16, 16), strides=(2, 2), padding='same',
                                     input_shape=[224, 224, 3])) #1
    model.add(layers.LeakyReLU())
    model.add(layers.Dropout(0.3))

    model.add(layers.Conv2D(128, (16, 16), strides=(2, 2), padding='same'))#,activation ='sigmoid'))
    model.add(layers.LeakyReLU())
    model.add(layers.Dropout(0.3))

    model.add(layers.Conv2D(256, (16, 16), strides=(2, 2), padding='same'))#,activation='tanh'))
    model.add(layers.LeakyReLU())
    model.add(layers.Dropout(0.3))

    model.add(layers.Conv2D(512, (16, 16), strides=(2, 2), padding='same'))#,activation='tanh'))
    model.add(layers.LeakyReLU())
    model.add(layers.Dropout(0.3))

    model.add(layers.Conv2D(256, (16, 16), strides=(2, 2), padding='same' ,activation='tanh'))
    model.add(layers.Dropout(0.3))

    model.add(layers.Flatten())
    model.add(layers.Dense(1))
    return model
#cross_entropy = tf.keras.losses.BinaryCrossentropy(from_logits=True)
cross_entropy = tf.keras.losses.MeanAbsoluteError()
def discriminator_loss(real_output, fake_output):
    real_loss = cross_entropy(tf.ones_like(real_output), real_output)
    fake_loss = cross_entropy(tf.zeros_like(fake_output), fake_output)
    total_loss = real_loss + fake_loss
    #print(real_loss,fake_loss,total_loss)
    return total_loss
def generator_loss(fake_output):
    print(cross_entropy(tf.ones_like(fake_output), fake_output))
    return cross_entropy(tf.ones_like(fake_output), fake_output)
generator_optimizer = tf.keras.optimizers.SGD(1e-4) #SGD
discriminator_optimizer = tf.keras.optimizers.SGD(1e-4) #SGD
EPOCHS = 10000
noise_dim = 100
num_examples_to_generate = 1
seed = tf.random.normal([num_examples_to_generate, noise_dim])
discriminator = make_discriminator_model()
decision = discriminator(generated_image)
print (decision)
@tf.function
def train_step(images):
    noise = tf.random.normal([rozmiar_probki, noise_dim])

    with tf.GradientTape() as gen_tape, tf.GradientTape() as disc_tape:
      generated_images = generator(noise, training=True)
      #generated_images = (generated_images - 255 ) / 255
      print(type(generated_images))
      real_output = discriminator(images, training=True)
      fake_output = discriminator(generated_images, training=True)

      gen_loss = generator_loss(fake_output)
      disc_loss = discriminator_loss(real_output, fake_output)

    gradients_of_generator = gen_tape.gradient(gen_loss, generator.trainable_variables)
    gradients_of_discriminator = disc_tape.gradient(disc_loss, discriminator.trainable_variables)

    generator_optimizer.apply_gradients(zip(gradients_of_generator, generator.trainable_variables))
    discriminator_optimizer.apply_gradients(zip(gradients_of_discriminator, discriminator.trainable_variables))
checkpoint_dir = 'content/drive/MyDrive/training_checkpoints'
checkpoint_prefix = os.path.join(checkpoint_dir, "ckpt")
checkpoint = tf.train.Checkpoint(generator_optimizer=generator_optimizer,
                                discriminator_optimizer=discriminator_optimizer,
                                 generator=generator,
                                 discriminator=discriminator)
def generate_and_save_images(model, epoch, test_input):
  predictions = model(test_input, training=False)
  decision = discriminator(predictions)
  print(decision)
  fig = plt.figure(figsize=(7, 7))

  for i in range(predictions.shape[0]):
      plt.subplot(1, 1, i+1)
      plt.imshow(predictions[i, :, :, 0] * 127.5 + 127.5, cmap='gray') #cmap='viridis') #cmap='grey')
      plt.axis('off')

  #plt.savefig("/content/drive/MyDrive/Mushroom/image_at_epoch_{:04d}.png".format(epoch))
  plt.show()
def train(dataset, epochs):
  for epoch in range(epochs):
    start = time.time()
    for image_batch in dataset:
      train_step(image_batch)

    # Produce images for the GIF as you go
    display.clear_output(wait=True)
    generate_and_save_images(generator,
                             epoch+1,
                             seed)

    # Save the model every 15 epochs
    if (epoch + 1) % 15 == 0:
      checkpoint.save(file_prefix = checkpoint_prefix)
      checkpoint.restore(tf.train.latest_checkpoint(checkpoint_dir))
    print ('Time for epoch {} is {} sec'.format(epoch + 1, time.time()-start))

  # Generate after the final epoch
  display.clear_output(wait=True)
  generate_and_save_images(generator,
                           epochs,
                           seed)
train(train_ds,EPOCHS) #train(z_ucz,EPOCHS) train_dataset
```

```python
#Sieć GAN ze zmienioną drastycznie strukturą sieci.
import tensorflow as tf
import glob
import imageio
import matplotlib.pyplot as plt
import numpy as np
import os
import PIL
from tensorflow.keras import layers
import time
from IPython import display
dane = "/content/drive/MyDrive/Mushrooms/Amanita"
tf.config.list_physical_devices('GPU')
#tf.debugging.set_log_device_placement(True)
s_px = 196
w_px = 196
rozmiar_probki = 150 #36
BUFFER_SIZE = 6000
m = 49 #7
n = 196 #28
s = 98 #14
r = 3 #1
#64
#(train_images, train_labels), (_, _) = tf.keras.datasets.mnist.load_data()
#train_images = train_images.reshape(train_images.shape[0], 28, 28, 1).astype('float32')
#train_images = (train_images - 127.5) / 127.5
#train_dataset = tf.data.Dataset.from_tensor_slices(train_images).shuffle(BUFFER_SIZE).batch(rozmiar_probki)
z_ucz = tf.keras.preprocessing.image_dataset_from_directory(dane,seed=66, image_size=(s_px,w_px),batch_size = rozmiar_probki,labels=None) #color_mode='grayscale')
def normalize(image):
  return (image - 255)/255
z_ucz = z_ucz.map(normalize)
#z_ucz = (z_ucz - 255)/255
#z_test = tf.keras.preprocessing.image_dataset_from_directory(dane, validation_split=0.1, subset="validation", seed=66, image_size=(s_px,w_px), batch_size=rozmiar_probki)
def make_generator_model():
    model = tf.keras.Sequential()
    model.add(layers.Dense(m*m*196, use_bias=False, input_shape=(100,)))
    model.add(layers.BatchNormalization())
    model.add(layers.LeakyReLU())

    model.add(layers.Reshape((m, m, 196)))
    assert model.output_shape == (None, m, m, 196)
    model.add(layers.Conv2DTranspose(196, (16, 16), strides=(1, 1), padding='same', use_bias=False))
    assert model.output_shape == (None, m, m, 196)
    model.add(layers.BatchNormalization())
    model.add(layers.Dropout(0.3))
    model.add(layers.LeakyReLU())

    model.add(layers.Reshape((m, m, 196)))
    assert model.output_shape == (None, m, m, 196)
    model.add(layers.Conv2DTranspose(196, (16, 16), strides=(1, 1), padding='same', use_bias=False,activation='sigmoid'))
    assert model.output_shape == (None, m, m, 196)
    model.add(layers.BatchNormalization())
    model.add(layers.Dropout(0.3))
    model.add(layers.LeakyReLU())

    model.add(layers.Conv2DTranspose(196, (5, 5), strides=(2, 2), padding='same', use_bias=False))
    assert model.output_shape == (None, s, s, 196)
    model.add(layers.BatchNormalization())
    model.add(layers.Dropout(0.3))
    model.add(layers.LeakyReLU())

    model.add(layers.Conv2DTranspose(r, (16, 16), strides=(2, 2), padding='same', use_bias=False, activation='tanh'))
    assert model.output_shape == (None, n, n, r)
    return model
generator = make_generator_model()
noise = tf.random.normal([1, 100])
generated_image = generator(noise, training=False)
plt.imshow(generated_image[0, :, :, 0], cmap='gray')
def make_discriminator_model():
    model = tf.keras.Sequential()
    model.add(layers.Conv2D(64, (5, 5), strides=(2, 2), padding='same',
                                     input_shape=[n, n, r]))
    model.add(layers.LeakyReLU())
    model.add(layers.Dropout(0.3))

    model.add(layers.Conv2D(128, (5, 5), strides=(2, 2), padding='same',activation ='sigmoid'))
    model.add(layers.LeakyReLU())
    model.add(layers.Dropout(0.3))

    model.add(layers.Conv2D(256, (5, 5), strides=(2, 2), padding='same',activation='tanh'))
    model.add(layers.LeakyReLU())
    model.add(layers.Dropout(0.3))

    model.add(layers.Flatten())
    model.add(layers.Dense(1))
    return model
#cross_entropy = tf.keras.losses.BinaryCrossentropy(from_logits=True)
cross_entropy = tf.keras.losses.MeanAbsoluteError()
def discriminator_loss(real_output, fake_output):
    real_loss = cross_entropy(tf.ones_like(real_output), real_output)
    fake_loss = cross_entropy(tf.zeros_like(fake_output), fake_output)
    total_loss = real_loss + fake_loss
    #print(real_loss,fake_loss,total_loss)
    return total_loss
def generator_loss(fake_output):
    print(cross_entropy(tf.ones_like(fake_output), fake_output))
    return cross_entropy(tf.ones_like(fake_output), fake_output)
generator_optimizer = tf.keras.optimizers.Nadam(1e-4)
discriminator_optimizer = tf.keras.optimizers.Nadam(1e-4)
EPOCHS = 1000
noise_dim = 100
num_examples_to_generate = 1
seed = tf.random.normal([num_examples_to_generate, noise_dim])
discriminator = make_discriminator_model()
decision = discriminator(generated_image)
print (decision)
@tf.function
def train_step(images):
    noise = tf.random.normal([rozmiar_probki, noise_dim])

    with tf.GradientTape() as gen_tape, tf.GradientTape() as disc_tape:
      generated_images = generator(noise, training=True)
      #generated_images = (generated_images - 255 ) / 255
      print(type(generated_images))
      real_output = discriminator(images, training=True)
      fake_output = discriminator(generated_images, training=True)

      gen_loss = generator_loss(fake_output)
      disc_loss = discriminator_loss(real_output, fake_output)

    gradients_of_generator = gen_tape.gradient(gen_loss, generator.trainable_variables)
    gradients_of_discriminator = disc_tape.gradient(disc_loss, discriminator.trainable_variables)

    generator_optimizer.apply_gradients(zip(gradients_of_generator, generator.trainable_variables))
    discriminator_optimizer.apply_gradients(zip(gradients_of_discriminator, discriminator.trainable_variables))
checkpoint_dir = 'content/drive/MyDrive/training_checkpoints'
checkpoint_prefix = os.path.join(checkpoint_dir, "ckpt")
checkpoint = tf.train.Checkpoint(generator_optimizer=generator_optimizer,
                                 discriminator_optimizer=discriminator_optimizer,
                                 generator=generator,
                                 discriminator=discriminator)
def generate_and_save_images(model, epoch, test_input):
  predictions = model(test_input, training=False)
  decision = discriminator(predictions)
  print(decision)
  fig = plt.figure(figsize=(5, 5))

  for i in range(predictions.shape[0]):
      plt.subplot(1, 1, i+1)
      plt.imshow(predictions[i, :, :, 0] * 255 + 255, cmap='viridis') #cmap='gray')
      plt.axis('off')
      #m = np.argmax(predictions[i],axis=1)
      #n = np.argmin(predictions[i],axis=1)
      #final_max = [predictions for i in m]
      #final_min = [predictions for i in n]
      #print(f"{final_min}   ")
      #print(f"{final_max}   ")
  #plt.savefig("/content/drive/MyDrive/Mushroom/image_at_epoch_{:04d}.png".format(epoch))
  plt.show()
def train(dataset, epochs):
  for epoch in range(epochs):
    start = time.time()
    for image_batch in dataset:
      train_step(image_batch)

    # Produce images for the GIF as you go
    display.clear_output(wait=True)
    generate_and_save_images(generator,
                             epoch+1,
                             seed)

    # Save the model every 15 epochs
    #if (epoch + 1) % 15 == 0:
      #checkpoint.save(file_prefix = checkpoint_prefix)
      #checkpoint.restore(tf.train.latest_checkpoint(checkpoint_dir))
    print ('Time for epoch {} is {} sec'.format(epoch + 1, time.time()-start))

  # Generate after the final epoch
  display.clear_output(wait=True)
  generate_and_save_images(generator,
                           epochs,
                           seed)
train(z_ucz,EPOCHS)
```

```python
import time
import pygame
import random
k = 3
#yg = (154,205,50)
black = (0,0,0)
ds = 300
class Cell():
    def __init__(self):
        self.gene = list()
        self.i = False
        self.l = True
    #def __del__(self):
    def gene_initalization(self):
        for i in range (0,8):
            self.gene.append(random.randint(0,1))
    def mutation(self):
        c = False
        a = random.randint(0, 100)
        b = (len(self.gene)-1)
        x = random.randint(0,b)
        if a >= 90 :
            if self.gene[x] == 0:
                self.gene[x] = 1
                c = True
            if self.gene[x] == 1 and c == False:
                self.gene[x] = 0
            print(self.gene[x],x)
        #if a > 95:
            #self.gene.append(random.randint(0,1))
        print(self.gene)
    def survival(self):
        d = []
        for i in range(len(self.gene)):
            d += (str(self.gene[i]))
        d = ''.join(d)
        d = int(d,2)
        if d >= 200:
            self.i = True
        if d <= 16:
            self.l = False
cells = list()
posy = list()
posx = list()
def life_emergence():
    for i in range(0,random.randint(5,12)):
        cells.append(Cell())
        cells[i].gene_initalization()
        posx.append(random.randint(2,298))
        posy.append(random.randint(2,298))
def cycle():
    for i in range(len(cells)):
        a = random.randint(1,8)
        x = random.randint(1,3)
        y = random.randint(1,3)
        if a == 1:
            posx[i] += x
        if a == 2:
            posy[i] += y
        if a == 3:
            posx[i] -= x
        if a == 4:
            posy[i] -= y
        if a == 5:
            posx[i] += x
            posy[i] += y
        if a == 6:
            posx[i] -= x
            posy[i] -= y
        if a == 7:
            posx[i] -= x
            posy[i] += y
        if a == 8:
            posx[i] += x
            posy[i] -= y
        cells[i].mutation()
        cells[i].survival()
        #time.sleep(0.001)
pygame.init()
dis = pygame.display.set_mode((ds, ds))
pygame.display.update()
pygame.display.set_caption('Cell project v 0.0.1')
life_emergence()
game_over = False
while not game_over:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            game_over = True
        if len(cells) <= 1:
            game_over = True
    cycle()
    for i in range(len(cells)):
        if cells[i].l == False and cells[i].i == False:
            del cells[i]
            del posx[i]
            del posy[i]
            break
    for i in range(len(cells)):
        a = random.randint(0, 50)
        for j in range(len(cells)-1):
            if ((posx[j]==(posx[j-1]-1)) or (posx[j]==(posx[j+1]+1)) or posy[j]==(posy[j-1]-1) or posy[j]==(posy[j+1]+1) or posx[j]==(posx[j-1]+1)) or (posx[j]==(posx[j+1]-1)) or posy[j]==(posy[j+1]-1) or posy[j]==(posy[j-1]+1 and (len(cells)>1) and a >= 45):
                x = cells[j]
                print(j)
                cells.append(x)
                print(cells)
                #print(cells[j].gene)
                if (posx[j]==posx[j-1]-1) or (posx[j]==(posx[j-1]+1)):
                    f = cells[j].gene[3:7]
                    r = cells[j-1].gene[0:4]
                    g = (r + f)
                    print(g)
                    cells[-1].gene = g
                if (posx[j] == posx[j + 1] - 1) or (posx[j] == (posx[j + 1] + 1)):
                    f = cells[j].gene[3:7]
                    r = cells[j + 1].gene[0:4]
                    g = (f+r)
                    print(g)
                    cells[-1].gene = g
                if (posy[j] == posy[j - 1] - 1) or (posy[j] == (posy[j - 1] + 1)):
                    f = cells[j].gene[3: 7]
                    r = cells[j - 1].gene[0: 4]
                    g = (f+r)
                    print(g)
                    cells[-1].gene = g
                if (posx[j] == posx[j + 1] - 1) or (posx[j] == (posx[j + 1] + 1)):
                    f = cells[j].gene[3:7]
                    r = cells[j + 1].gene[0: 4]
                    g = (f+r)
                    print(g)
                    cells[-1].gene = g
                posy.append(posy[j]-15)
                posx.append(posx[j]+15)
            break
    dis.fill(black)
    for i in range(len(cells)):
        d = " "
        for j in range(len(cells[i].gene)):
            d += (str(cells[i].gene[j]))
            print(j)
            print(d)
        d = ''.join(map(str,d))
        d = int(d,2)
        color = (d,d,d)
        #print(color)
        pygame.draw.rect(dis,color, [posx[i], posy[i], k, k])
        if posy[i] >= ds:
            posy[i] -= 15
        elif posy[i] < 0:
            posy[i] += 15
        if posx[i] >= ds:
            posx[i] -= 15
        elif posx[i] < 0:
            posx[i] += 15
    pygame.display.update()
pygame.quit()
quit()
```
